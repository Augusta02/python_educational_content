```python
# Question 1
# Write a Python program that accepts the user's first and last name, and prints them in reverse order with a space between them.


first_name = str(input('Enter firstname: '))
last_name = str(input('Enter lastname: '))


print(last_name + " " + first_name)
```

```python
# Question 2
# Write a Python program that accepts an integer (n) and computes the value of n+nn+nnn.

x = int(input('Enter value: '))

n1 = int('%s' % x)
n2 = int('%s%s' % (x,x))
n3 = int('%s%s%s' % (x,x,x))
result = n1 + n2 +n3
print(result)
```

```python
# Question 3
# Write a Python program to find whether a given number (accept from the user) is even or odd, and print out an appropriate message to the user.

y = int(input('Enter value: '))

if y % 2 != 0:
  print('{} is a odd number'. format(y))
else:
  print('{} is an even number'.format(y))
```

```python
# Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5,
#  between 2000 and 3200 (both included). The numbers obtained should be printed in a sequence on a single line.
# Hints: Consider use range(#begin, #end) method 

result = []

for i in range(2000, 3201):
  if (i % 7 == 0) and (i % 5 != 0):
    result.append(i)
  print(result)
```

```python
# Write a program that can compute the factorial of a given number.
#  The results should be printed in a sequence on a single line.
#  Suppose the following input is supplied to the program: 8. Then, the output should be: 40320 

# Hint: The factorial function (symbol: !) multiplies all whole numbers from our chosen number down to 
number = int(input('Enter value:'))
result = 1

for i in range(1, number+1):
  result = result * i
print(result)
```
